# Load the data files and change the column names if needed

```python
raw_content_data = pd.read_csv('Content.csv',index_col=0)
raw_content_data
```

```python
raw_content_data.columns=['Content_ID', 'User_ID', 'Content_type', 'Category', 'URL']
raw_content_data
```

```python
raw_reaction_data = pd.read_csv('Reactions.csv',index_col=0)
raw_reaction_data
```

```python
raw_reaction_data.columns=['Content_ID', 'User_ID', 'Reaction_type', 'DateTime']
raw_reaction_data
```

```python
raw_reaction_type_data = pd.read_csv('ReactionTypes.csv',index_col=0)
raw_reaction_type_data
```

```python
raw_reaction_type_data.columns=['Reaction_type', 'Sentiment', 'Score']
raw_reaction_type_data
```

# Task1 Removing columns which are not relevant to this task.

```python
content_data_rc = raw_content_data.copy()
content_data_rc = content_data_rc.drop(['URL','User_ID'], axis=1)
reaction_data_rc = raw_reaction_data.copy()
reaction_data_rc=reaction_data_rc.drop(['User_ID'],axis=1)
reaction_type_data_rc = raw_reaction_type_data.copy()
```

```python
content_data_rc
```

```python
reaction_data_rc
```

# Task2) Change the data type of some values within a column

```python
content_data_cd = content_data_rc.copy()
reaction_data_cd= reaction_data_rc.copy()
reaction_type_data_cd = reaction_type_data_rc.copy()
```

### There are some values in Category column from the content dataset which have quotation marks and those should be removed because the names should be the same for each category so that it will count the scores later correctly

```python
# Find the ones with quotation marks and remove them
for i in range(content_data_cd.shape[0]):
    if content_data_cd['Category'][i].find('\"') == 0:
        content_data_cd['Category'][i] = content_data_cd['Category'][i].replace('"', '')
        
# if it's printed something other than -1, that means that still has quotation mark.
for i in range(content_data_cd.shape[0]):
    print(content_data_cd['Category'][i].find('\"'))
```

# Task3) Removing rows that have values which are missing

```python
content_data_rr = content_data_cd.copy()
reaction_data_rr = reaction_data_cd.copy()
reaction_type_data_rr = reaction_type_data_cd.copy()
```

## 1) Check the total number of raw data sets

```python
print(content_data_rr.shape)
print('number of raw content data:', content_data_rr.shape[0] )

print(reaction_data_rr.shape)
print('number of raw reaction data:', reaction_data_rr.shape[0] )

print(reaction_type_data_rr.shape)
print('number of raw content data:', reaction_type_data_rr.shape[0] )
```

## 2) Check how many missing values there are for each data set (Just to make sure and to have a look at which columns have missing values

```python
content_columns = content_data_rr.columns.values
reaction_columns = reaction_data_rr.columns.values
reaction_type_columns = reaction_type_data_rr.columns.values
```

```python
for i in range(len(content_columns)):
    print('The column ', content_columns[i], ' in the content dataset has ', raw_content_data[content_columns[i]].isnull().sum(), ' missing values')
print('\n')    
for i in range(len(reaction_columns)):
    print('The column ', reaction_columns[i], 'in the reaction dataset has ', raw_reaction_data[reaction_columns[i]].isnull().sum(), ' missing values')
print('\n')    
for i in range(len(reaction_type_columns)):
    print('The column ', reaction_type_columns[i], ' in the reaction type dataset has ', raw_reaction_type_data[reaction_type_columns[i]].isnull().sum(), ' missing values')
print('\n')    
```

## 3) Remove the rows with missing values
## In content dataset: no rows with missing values
## In reaction dataset: rows with missing values in the User ID and Type columns
## In reaction type dataset: no rows with missing values

```python
content_data_rr = content_data_rr.dropna(axis=0)

for i in range(len(content_columns)):
    print('The column ', content_columns[i], ' in the content dataset has ', content_data_rr[content_columns[i]].isnull().sum(), ' missing values')
```

```python
reaction_data_rr = reaction_data_rr.dropna(axis=0)

for i in range(len(reaction_columns)):
    print('The column ', reaction_columns[i], 'in the reaction dataset has ', reaction_data_rr[reaction_columns[i]].isnull().sum(), ' missing values')
```

```python
reaction_type_data_rr = reaction_type_data_rr
```

```python
print(content_data_rr.shape)
print('number of content data:', content_data_rr.shape[0] )

print(reaction_data_rr.shape)
print('number of reaction data:', reaction_data_rr.shape[0] )

print(reaction_type_data_rr.shape)
print('number of content data:', reaction_type_data_rr.shape[0] )
```

```python
final_content_data=content_data_rr.copy()
final_reaction_data=reaction_data_rr.copy()
final_reaction_type_data=reaction_type_data_rr.copy()
```

## Save the final datasets into new csv files

```python
final_content_data.to_csv('final_content_data.csv',index=False,header=True)
final_reaction_data.to_csv('final_reaction_data.csv',index=False,header=True)
final_reaction_type_data.to_csv('final_reaction_type_data.csv',index=False,header=True)
```
